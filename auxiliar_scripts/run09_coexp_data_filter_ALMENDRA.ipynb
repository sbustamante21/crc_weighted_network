```r
library(reshape2)
library(reshape)
library(tidyverse)
library(data.table)
library(stringr)
```

```r
#load data obtained in run08_WGCNA_coexpanalysis.ipynb

module_membership <- read.delim("your_path/geneModuleMembership.csv", sep ='\t', header=TRUE)
module_significance <- read.delim("your_path/PvaluegeneModuleMembership.csv", sep='\t', header= TRUE) 
```

```r
#create geneid column
module_membership$geneid<-rownames(module_membership)
module_significance$geneid<-rownames(module_significance)
```

```r
mm_melted<-reshape2::melt(module_membership,id.vars = "geneid")
```

```r
colnames(mm_melted)<-c("geneid","module","mm")
```

```r
mmsig_melted<-reshape2::melt(module_significance,id.vars = "geneid")
```

```r
colnames(mmsig_melted)<-c("geneid","module","p_mm")
```

```r
mmsig_melted$module <- gsub("p.", "", mmsig_melted$module, fixed = TRUE)
```

```r
#merge data
merged_mm <- merge(mm_melted, mmsig_melted, by = c("geneid","module"))
```

```r
colnames (merged_mm)
```

```r
#Filter by your criteria, in this case: 

filtered_mm<-subset(merged_mm, abs(mm) >= 0.8 & p_mm < 0.00001)


write.table(filtered_mm, "your_path/filtered_mm.tsv",sep = '\t',row.names = FALSE,quote = FALSE)
```

```r
#Load gene significance (GS) data for each trait analyzed in WGCNA

#GS RESISTANT
gs_resistant<-read.delim("your_path/geneTraitSignificance_resistant.csv",sep = '\t', header=TRUE)
gsp_resistant<-read.delim("your_path/GeneSignificancePvalue_resistant.csv", sep = '\t', header=TRUE)

#generate geneid
gs_resistant$geneid<-rownames(gs_resistant)
gsp_resistant$geneid<-rownames(gsp_resistant)

#merge: GS data and p-value
merged_gs_resistant <- merge(gs_resistant,gsp_resistant, by = "geneid")

#Filter by your criteria, in this case: 
filtered_resistant<-subset(merged_gs_resistant, GS.Resistant >= 0.5 & p.GS.Resistant < 0.01)

#merge data and add module results (filtered_mm) for resistant genes
resistant_genes <- merge(filtered_resistant,filtered_mm, by = "geneid")

#save
write.table(resistant_genes, "your_path/resistant_genes_filtered.tsv",sep = '\t',row.names = FALSE,quote = FALSE)
```

```r
#GS SUSCEPTIBLE
gs_susceptible <-read.delim("your_path/geneTraitSignificance_susceptible.csv",sep = '\t', header=TRUE)
gsp_susceptible <-read.delim("your_path/GeneSignificancePvalue_susceptible.csv", sep = '\t', header=TRUE)

#generate geneid
gs_susceptible$geneid<-rownames(gs_susceptible)
gsp_susceptible$geneid<-rownames(gsp_susceptible)

#merge: GS data and p-value
merged_gs_susceptible <- merge(gs_susceptible, gsp_susceptible, by = "geneid")

#Filter by your criteria, in this case: 
filtered_susceptible <- subset(merged_gs_susceptible, GS.Susceptible >= 0.5 & p.GS.Susceptible < 0.01)

#merge data and add module results (filtered_mm) for resistant genes
susceptible_genes <- merge(filtered_susceptible, filtered_mm, by = "geneid")

#save
write.table(susceptible_genes, "your_path/susceptible_genes_filtrados.tsv",sep = '\t',row.names = FALSE,quote = FALSE)
```
